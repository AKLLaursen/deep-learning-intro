# Examples of Style Transfer

This notebooks contains various examples of Style transfer in the style of Johnson et. al. *Perceptual losses for real-time style transfer.*

## Setup

Load functions, libraries and set paths.

```python
# Plots displayed inline in notebook
%matplotlib inline

# Make help libraries available
import sys

sys.path.append('D:/anlaursen/libraries')

# Set visible devices, so as to just use a single GPU.
import os
os.environ["CUDA_DEVICE_ORDER"] = "PCI_BUS_ID"   # see issue #152
os.environ["CUDA_VISIBLE_DEVICES"] = "0"
```

```python
import bcolz
import tensorflow as tf

from styletransfer.styletransfer import StyleTransfer

#tf.GPUOptions(per_process_gpu_memory_fraction = 0.333) 
```

Load training data can convert to numpy array

```python
imagenet = bcolz.open('data/trn_resized_288.bc')[:]
```

## Van Gogh

Attempt using a Van Gogh self portrait

Create style transfer model

```python
van_gogh_style = StyleTransfer(style_path = 'data/styles/van-gogh-self-portrait.jpg')
```

Display style image

```python
van_gogh_style.plot_style()
```

Compile model

```python
van_gogh_style.compile()
```

Train model

```python
van_gogh_style.fit(imagenet,
                   batch_size = 8,
                   epochs = 2,
                   learning_rate = 1e-3)
```

Lower learning rate and run another epoch

```python
van_gogh_style.fit(imagenet,
                   batch_size = 8,
                   epochs = 1,
                   learning_rate = 1e-4)
```

Save weights

```python
van_gogh_style.m_style.save_weights('results/van-gogh-self-portrait.h5')
```

Plot test image.

```python
van_gogh_style.plot_image('test/akll-bw-big.jpg')
```

```python
van_gogh_style.plot_image('test/akll-color-big.jpg')
```

Initiate style transfer model

```python
van_gogh_style.get_top()
```

Save weights

```python
van_gogh_style.top_model.save_weights('results/van-gogh-self-portrait-top.h5')
```

Make prediction

```python
van_gogh_akll_big = van_gogh_style.predict('test/akll-bw-big.jpg')
```

```python
van_gogh_akll_big = van_gogh_style.predict('test/akll-color-big.jpg')
```
